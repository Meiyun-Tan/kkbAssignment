```python
def facetgrid_plot(feature_name, hue='label'):
    """绘制特征在不同性别下的核密度曲线"""
    sns.set_palette(palette='hsv')
    facet = sns.FacetGrid(df, hue=hue, aspect=4)   # 创建用于绘制条件关系的多图网格
    facet.map(sns.kdeplot, feature_name, shade=True)     # 拟合并绘制单变量核密度估计图。
    facet.set(xlim=(0, df[feature_name].max()))   # 设置x轴刻度范围0~最大值
    facet.add_legend()    #  添加图例
    plt.show()
```

## 2.1 多重共线性分析
计算方差膨胀系数

```python
def get_variance_inflation_factor(df_new):
    """计算df中特征f与其他特征间的方差膨胀系数"""
    varianceif = {}
    for f in df_new.columns:
        exog = df_new.values
        exog_idx = df_new.columns.get_loc(f)
        varianceif[f] = variance_inflation_factor(exog, exog_idx)
        varianceif = dict(sorted(varianceif.items(), key=lambda x: x[1], reverse=True))
    return varianceif
```

# 5 总结：

1. 对数据集建模的第一版代码中，我尝试了不做特征工程，直接丢给模型去预测。本代码是第三版，与第一版相比，模型的准确率除了支持向量机和GBDT+LR以外，其他均有所上升。
2. 从对测试集的准确率上看，xgboost模型得到的准确率最高：98.74%；
3. 尝试了利用方差膨胀系数对特征进行筛选，在删除特征时，需要逐个删除查看效果，以便找到更好的效果；
4. 尝试了对特征meanfun的离散化，离散化后，LR模型准确率终于升到98%；
5. 对GBDT+LR模型而言，特征工程后，准确率反而有所下降，推测GBDT在这里的作用是降维，而我删除的特征太多，所以导致GBDT的作用降低；
6. 线性特征多的数据集使用线性SVM进行预测，准确率不一定比其他支持向量机模型高；
7. 特征工程的处理，应该还可以有更多方法，暂时还有待挖掘。

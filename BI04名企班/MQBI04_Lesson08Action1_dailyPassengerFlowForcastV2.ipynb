交通流量预测     
JetRail高铁的乘客数量预测    
数据集：jetrail.csv    
根据过往3年的数据（2012 年 8 月至 2014 年 9月），需要用这些数据预测接下来 7 个月的乘客数量

## 1.3 查看序列情况

### 1.3.1 可视化查看时间序列概况

```python
result = tsa.seasonal_decompose(x=df_day)
```

```python
fig = result.plot()
# 修改图像的高度和宽度
fig.set_figheight(8)
fig.set_figwidth(16)
plt.show()
```

### 1.3.2 时间序列的ADF检验（单位根检验）

```python
adf = tsa.stattools.adfuller(df_day)
```

```python
type(adf), len(adf)
```

```python
print(f"adf检验统计量为:{adf[0]:.4f}")
print(f"adf检验Pvalue为:{adf[1]:.4f}")
print(f"结果使用的延迟阶数:{adf[2]}")
print(f"ADF回归和计算临界值所使用的观察次数:{adf[3]}")
print(
    f"99%置信度的临界值为:{adf[4]['1%']:.4f} \n95%置信度的临界值为:{adf[4]['5%']:.4f} \n90%置信度的临界值为:{adf[4]['10%']:.4f}"
)
print(f"最大的信息准则AIC的值{adf[5]:.4f}")
```

- **确定该序列能否平稳主要看以下几点：**
    - 1%、%5、%10不同程度拒绝原假设的统计值和ADF Test result的比较，ADF Test result同时小于1%、5%、10%即说明非常好地拒绝该假设；本数据中，adf结果为2.9864， 大于1%、5%和10%的level统计值，说明序列不平稳。
    - P-value是否非常接近0，本数据中，P-value 为 1，大于0，也可以说明序列不平稳
    - ADF检验的原假设是存在单位根，只要这个统计值是小于1%水平下的数字就可以极显著的拒绝原假设，认为数据平稳。注意，ADF值一般是负的，也有正的，但是它只有小于1%水平下的才能认为是及其显著的拒绝原假设。
    - 对于ADF结果在1% 以上 5%以下的结果，也不能说不平稳，关键看检验要求是什么样子的。
- ADF检验只适合AR(P)模型，当检验结果是平稳序列时，则适用于 AR模型。

#### 3.1.2.1 认识预测结果DataFrame

```python
future.columns
```

- daily_seasonality=False的加法模型，有：
     1. `forecast['additive_terms'] = forecast['weekly'] + forecast['yearly']`；
     1. `forecast['yhat']  = forecast['trend'] +  forecast['additive_terms']` 
     1. `forecast['yhat']  = forecast['trend'] +forecast['weekly'] + forecast['yearly']`
- 加法模型，所以'multiplicative_terms', 'multiplicative_terms_lower', 'multiplicative_terms_upper'这3列为空

```python
# 验证第一条，结果十分接近于0，所以第一条成立
(future['additive_terms'] - (future['weekly'] + future['yearly'])).sum()
```

```python
# 验证第二条，结果为0，所以第二条成立
(future['yhat'] - (future['trend'] + future['additive_terms'])).sum()
```

```python
# 验证第三条，结果极小、接近于0，所以第三条成立
(future['yhat'] -
 (future['trend'] + future['weekly'] + future['yearly'])).sum()
```

#### 3.1.2.2 可视化预测结果

```python
fig = model.plot(future)
fig.set_figwidth(12)
fig.set_figheight(6)
# 绘制突变点位置，默认均匀分布在前 80% 的时间序列数据中
a = add_changepoints_to_plot(fig.gca(), model, future)
plt.show()
```

浅蓝色部分是置信区间，浅蓝色的上边界代表预测的置信区间的上边界yhat_upper，浅蓝色的下边界代表预测的置信区间的下边界值yhat_lower；黑色代表真实值；深蓝色的代表预测的平均值。

未来7个月，每日乘客数量仍然会呈整体上升趋势；15年3月底、4月初会略降低，然后继续增长。2015年4月底预计每日乘客数约在12000-14000范围内。

#### 3.1.2.3 查看预测的成分分析
展示预测中的趋势、周效应和年度效应

```python
fig = model.plot_components(future)
fig.set_figheight(6)
fig.set_figwidth(12)
plt.show()
```

```python
# 手动绘制趋势、周效应和年度效应
figs, axes = plt.subplots(3, 1, figsize=(12, 6))
axes[0].plot(future.set_index('ds')['trend'])
axes[1].plot(future.set_index('ds')['weekly'][1:8])
axes[2].plot(future.set_index('ds').loc['2013', 'yearly'])

weekname = future.set_index('ds').index.day_name()

axes[1].set_xticklabels(weekname[1:8])
for i, y in enumerate(["trend", "weekly", "yearly"]):
    axes[i].set_ylabel(y)
plt.show()
```

可见，plot_components绘制的结果，是从future的trend、weekly、yearly三列的取值绘制而得。

预测结果成分分析解读：
- 第一个子图trend趋势图：显示数据呈上升趋势，14年后曲线斜率增加，说明增长速度更快；
- 第二个子图weekly周效应图：显示周末两天客流量减少，周一到周四增加；weekly的值为`x`，表示在trend的基础上，加`x`；x为正数时增加，为负数时表示在trend的基础上减少。因此，weekly这条线的高低反应了客流量的趋势。
- 第三个子图yearly年度效应：1月-3月乘客流量会有个波峰，应该是春节的缘故；9月~11月为全年客流量高峰。

```python
# 2012年到13年9月和10月每日乘客流量变动曲线
_, axes = plt.subplots(3, 1, figsize=(12, 12))
axes[0].plot(df_day.set_index('ds').loc['2012-09':'2012-10'],
             color='b')
axes[1].plot(df_day.set_index('ds').loc['2013-09':'2013-10'],
             color='g')
for i, y in enumerate(["2012", "2013"]):
    axes[i].set_title("%s年9-10月每日高铁客流量变动曲线" % y)
    
axes[2].plot(future.set_index('ds').loc['2014-09':'2014-10']['yhat'], color='purple')
axes[2].set_title("预测的2014年9-10月每日高铁客流量变动曲线")
plt.show()
```

可以看到，2012年13年10月1日和8日是国庆期间的客流量最高峰；14年的预测结果中也预测出了10月1日和8日是客流量高峰。

## 3.2 带假期建模


### 3.2.1 准备假期数据

```python
start = df_day['ds'][0]
end = datetime(2015, 4, 25)
start, end
```

```python
# 获得2012-8-25到2014-9-25中国假期的日期
holidays = get_holidays(start=start, end=end, include_weekends=False)
```

```python
type(holidays), len(holidays), type(holidays[0])
```

```python
get_holiday_detail(holidays[1])
```

```python
holidays_name = [get_holiday_detail(day)[1] for day in holidays]
```

```python
# 保留字段：节假日 holiday 、日期戳 ds、lower_window 以及upper_window
# 将节日看成是一个正态分布，把活动期间当做波峰，lower_window 以及upper_window 的窗口作为扩散
# 国庆和元旦假期肯定休息时间不一致，lower_window 以及upper_window可以理解为假日延长时限，
holiday_df = pd.DataFrame({
    "ds": holidays,
    "holiday": holidays_name,
    "lower_window": -3,
    "upper_window":3
})
```

```python
holiday_df.tail(2)
```

### 3.2.2 带节假日建模

```python
# 带节假日建模
modelh = Prophet(holidays=holiday_df, yearly_seasonality=True, n_changepoints=50)
```

```python
# 训练集训练
modelh.fit(train_df)
```

```python
len(modelh.changepoints)
```

1. 当csv中日期格式是dd-mm-yyyy的时候，若在读入数据集的时候使用parse_dates参数解析日期列，解析的日期数据将会出错；
2. 输入模型的数据时间列名必须为ds，待预测的数据列名必须为y；
3. Prophet模型训练后，可以直接调用实例方法直接生成待预测的时间序列；
4. 可视化中:
    - 既可以对整个预测结果进行可视化`fig = model.plot()`
    - 也可以可视化所有成分`model.plot_components()`
    - 还可以指定仅可视化某单个成分`model.plot.plot_forecast_component()`；
    - 同时也探索了修改`fig = model.plot()`中figure图像大小的方法(1.3.1)：
        - 设置图像高度：`fig.set_figheight(8)`
        - 设置图像宽度：`fig.set_figwidth(16)`
    - 在所绘制的图像上想继续绘制时的方法：调用Line2D对象的axes的方法在该轴上继续绘制(Line2D.axes.plot())，具体见3.1.1.3 ；
5. 增加节假日、增加突变点个数，确实使得模型拟合程度增加了(表现为预测均值的抖动程度增加)；
6. 对测试集，我们不需要重复构建测试集的待预测数据，直接对测试集的DataFrame进行预测即可，结果一致。

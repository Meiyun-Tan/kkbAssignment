## 1.4 seasonal_decompose分解

```python
# 可视化指定区间内的数据
def plot_stl(data):
    # STL返回3个部分，趋势、季节、残差
    result = sm.tsa.seasonal_decompose(data, period=30)
    fig = result.plot()
    fig.set_figwidth(16)
    fig.set_figheight(8)
    plt.show()
    return result
```

### 1.4.1 拆解总购买量

```python
pur_result = plot_stl(purchase)
```

```python
red_result = plot_stl(redeem)
```

# 2 时间序列的ADF检验

```python
def print_adf(adf):
    print(f"adf检验统计量为:{adf[0]:.4f}")
    print(f"adf检验Pvalue为:{adf[1]:.4f}")
    print(f"结果使用的延迟阶数:{adf[2]}")
    print(f"ADF回归和计算临界值所使用的观察次数:{adf[3]}")
    print(
        f"99%置信度的临界值为:{adf[4]['1%']:.4f} \n95%置信度的临界值为:{adf[4]['5%']:.4f} \n90%置信度的临界值为:{adf[4]['10%']:.4f}"
    )
    print(f"最大的信息准则AIC的值{adf[5]:.4f}")
```

## 2.1 对总购买量序列进行ADF检验

```python
purchase.head(2)
```

```python
purchase2 = purchase[purchase.index >= '2014-04-01']
purchase2.shape
```

```python
_ = plot_stl(purchase2)
```

### 2.1.1  对差分前的purchase进行检验

```python
adf = sm.tsa.stattools.adfuller(purchase)

print_adf(adf)
```

- **确定该序列能否平稳主要看以下几点：**    
    - 1%、%5、%10不同程度拒绝原假设的统计值和ADF Test result的比较，ADF Test result同时小于1%、5%、10%即说明非常好地拒绝原假设；本数据中，adf结果为-1.5899， 大于1%、5%和10%的level统计值，说明无法拒绝原假设，序列不平稳。    
    - 本数据中，P-value 为0.4887 远大于0，因此不能拒绝原假设。    
    
ADF检验的原假设是存在单位根，只要这个统计值是小于1%水平下的数字就可以极显著的拒绝原假设，认为数据平稳。注意，ADF值一般是负的，也有正的，但是它只有小于1%水平下的才能认为是及其显著的拒绝原假设。    

对于ADF结果在1% 以上 5%以下的结果，也不能说不平稳，关键看检验要求是什么样子的。    

ADF检验只适合AR(P)模型，当检验结果是平稳序列时，则适用于 AR模型。

```python
purchase.diff(1)
```

### 2.1.2 一阶差分后的purchase的ADF检验值

```python
adf = sm.tsa.stattools.adfuller(purchase.diff(1).iloc[1:])

print_adf(adf)
```

进行一阶差分后，ADF Test result同时小于1%、5%、10%即说明非常好地拒绝原假设；一阶差分后的purchase的时间序列是平稳序列。

- 什么是自相关函数 autocorrelation function（ACF）？   
$X_t$ 与 $X_{t+k}$ 的自相关系数即（自协方差 / k=0时的自协方差）。ρk的公式：     

$$r _{k} = \frac{\sum _{t=1}^{N-k}(x_t - \bar{x})(x_{t+k}  - \bar{x})}{\sum _{t=1}^{N}(x_t - \bar{x})}$$     
 
$r_{0}$永远等于1，$r_{k}$称为样本自相关函数(ACF)         

- 自相关图 correlogram: 样本$x_t$ 与 $x_{t+k}$，把k从0到n的自相关系数求出并直线。 横坐标是lag，即k值。    
自相关图中的两条横虚线表示样本自相关函数的两个标准差的上下限。上图右下角的自相关图中，阴影部分是两个标准差的上下限范围内

**残差**在数理统计中是指实际观察值与估计值（拟合值）之间的差。“残差”蕴含了有关模型基本假设的重要信息。如果回归模型正确的话， 我们可以将残差看作误差的观测值。它应符合模型的假设条件，且具有误差的一些性质。利用残差所提供的信息，来考察模型假设的合理性及数据的可靠性称为残差分析。   

主要关切的是确保我们的模型的残差是不相关的，并且残差的均值为零。
- 左上角图：t的标准残差图
- 右上角图：N(1, 0)是正态分布的标准符号，平均值0 ，标准偏差为1。密度曲线KDE远高于正态分布，这说明残差分布不符合正态分布。
- 左下角图：qq图显示，残差（蓝点）的有序分布没有遵循采用N(0, 1)的标准正态分布采样的线性趋势。 这同样说明残差分布不符合正态分布。
- 右下角图：自相关图，在lag=5和lag=8时，有一定的相关性。  
模型的拟合性还不够好。
